```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
file3 = 'Resources/crime_data/Dallas.csv'
df = pd.read_csv(file3, encoding="ISO-8859-1")
df.head()
```

```python
df = df[[ "IncidentNum", "ArrestYr", "ArArrestTime",  "ArLCity", "DrugType",  "ClothingWorn"]].dropna()
df
```

```python
rds_connection_string = "postgres:boozer11@localhost:5432/police-violence-racial-equity"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
df.to_sql(name = "dallas", con = engine, if_exists = "append", index = False)
```

```python
pd.read_sql_query('select *  from dallas;',con=engine)
```
